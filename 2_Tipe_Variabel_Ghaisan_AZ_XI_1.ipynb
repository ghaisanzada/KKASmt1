##Tugas kelompok menghitung nilai volume bola

```python
r = float(input("Masukkan jari-jari bola (dalam satuan cm): "))
pi = 3.14
volume = (4/3) * pi * r**3

print(f"Volume bola dengan jari-jari {r} cm adalah {volume:.2f} cm続")
```

```python

r = float(input("Masukkan jari-jari bola (dalam satuan cm): "))
import math
volume = (4/3) * math.pi * r**3

print(f"Volume bola dengan jari-jari {r} cm adalah {volume:.2f} cm続")
```
